## Introduction to Scikit Learn (SKLearn)

Here we are going to cover the SkLearn python Machine Learning library.

Following is the Python Machine Learning Workflow:
    
0. End-to-End Machine Learning Framework Workflow
1. Getting the Data Ready
2. Choosing the right Estimator/Algorithm for the Data Acquired
3. Fitting the Model/Algorithm/Estimator 
4. Evaluating the Model
5. Improve the Model through Experimentation
6. Save and Reload the Trained Model
7. Putting it all Together

```python
# Getting the Data Ready
import pandas as pd
heart_disease = pd.read_csv('../data/heart-disease.csv')
heart_disease.head()
```

```python
# Feature matrix
X = heart_disease.drop(['target'], axis=1)
X.head()
```

```python
# Target column
y = heart_disease['target']
y.head()
```

```python
# 2. Choose a right model
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier()

clf.get_params()
```

```python
# 3. Fit the model on training data
from sklearn.model_selection import train_test_split
import numpy as np
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train
```

```python
X_test
```

```python
clf.fit(X_train, y_train)
```

```python
y_train
```

```python
# make predictions

y_pred = clf.predict(X_test)
y_pred
```

```python
# 4. Evaluate the model
clf.score(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
print(classification_report(y_test, y_pred))
```

```python
print(confusion_matrix(y_test, y_pred))
```

```python
print(accuracy_score(y_test, y_pred))
```

```python
# 5. Experiment with different model
np.random.seed(42)
for i in range(10, 100, 10):
    print(f"Model accuracy with n_estimators {i}")
    clf = RandomForestClassifier(n_estimators=i).fit(X_train, y_train)
    print(f"Model fitted with test accuracy of {clf.score(X_test, y_test) * 100:.2f}")
```

```python
# 6. Save and Reload model
import pickle

with open("random_forest_model.pkl", "wb") as file:
    pickle.dump(clf, file)
```

```python
with open("random_forest_model.pkl", "rb") as file:
    model = pickle.load(file)

model.score(X_test, y_test)
```
