## Matplotlib Example Workflow

```python
%matplotlib inline
import matplotlib.pyplot as plt

y = np.random.normal(size=(10))
x = np.linspace(1, 10, num=10)

fig, ax = plt.subplots(figsize=(6, 6))

ax.plot(x, y)

ax.set(title='Sample Normal Data',
      xlabel='Data Points',
      ylabel='Normal Data Points')

fig.savefig('../images/sample-normal-data.png')
```

```python
nut_prices = {'almond': 100,
             'peanut': 200,
             'cashew': 300}
fig, ax = plt.subplots()
ax.bar(nut_prices.keys(), nut_prices.values())
ax.set(title='Nut Prices',
      xlabel='Nuts',
      ylabel='Prices')
plt.show()
```

```python
a = np.random.randn(1000)
fig, ax = plt.subplots()
ax.barh(list(nut_prices.keys()), list(nut_prices.values()));
```

```python
fig, ax = plt.subplots()
ax.hist(a);
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10, 5))
ax1.plot(x, x**2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_prices.keys(), nut_prices.values());
ax4.hist(a);
```

```python
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10, 5))
ax[0, 0].plot(x, x**3);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_prices.keys(), nut_prices.values())
ax[1, 1].hist(np.random.randn(1000));
```

```python
# let's try on different dataset
heart_disease = pd.read_csv('../heart-disease.csv')
heart_disease
```

```python
heart_disease['age'].plot.hist()
```

```python
heart_disease.plot.hist(figsize=(10, 50), subplots=True);
```

```python
heart_disease.head()
```

```python
over_50 = heart_disease[heart_disease['age'] > 50]
over_50.head()
```

```python
# Pyplot api method
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             figsize=(10, 5));
```

```python
# OO Method
fig, ax = plt.subplots(figsize=(10, 5))
over_50.plot(kind='scatter',
             x='age',
             y='chol',
             c='target',
             ax=ax);

ax.set_xlim([45, 80]);
```

```python
## OO method simplified

fig, ax = plt.subplots(figsize=(10, 6))

scattey = ax.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'])

ax.set(title='Heart Disease Cholesterol Level by Age',
       xlabel='Age',
       ylabel='Cholesterol');

ax.legend(*scattey.legend_elements(), title='Target');

ax.axhline(y=over_50['chol'].mean(),
          linestyle='--',
          c='red');
```

## Subplot with Cholesterol, Age and Max Heart Rate

```python
fig, ax = plt.subplots(nrows=2,
                       ncols=1,
                       figsize=(10, 12),
                       sharex=True)

scatter_chol = ax[0].scatter(x=over_50['age'],
                             y=over_50['chol'],
                             c=over_50['target'])

ax[0].set(title='Heart Disease Cholesterol Level by Age',
          ylabel='Cholesterol')

ax[0].legend(*scatter_chol.legend_elements(), title='Target')

ax[0].axhline(y=over_50['chol'].mean(), linestyle='--', c='red');

##

scatter_thal = ax[1].scatter(x=over_50['age'],
                             y=over_50['thalach'],
                             c=over_50['target'])

ax[1].set(title='Heart Disease Max Heart Rate by Age',
          xlabel='Age',
          ylabel='Max Heart Rate')

ax[1].legend(*scatter_thal.legend_elements(), title='Target')

ax[1].axhline(y=over_50['thalach'].mean(), linestyle=':', c='blue');

fig.suptitle('Heart Disease Plot', fontsize=16, fontweight='bold');
```
