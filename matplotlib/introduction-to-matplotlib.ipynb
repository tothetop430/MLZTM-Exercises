## Matplotlib Example Workflow

```python
%matplotlib inline
import matplotlib.pyplot as plt

y = np.random.normal(size=(10))
x = np.linspace(1, 10, num=10)

fig, ax = plt.subplots(figsize=(6, 6))

ax.plot(x, y)

ax.set(title='Sample Normal Data',
      xlabel='Data Points',
      ylabel='Normal Data Points')

fig.savefig('../images/sample-normal-data.png')
```

## Different types of plots
* Scatter
* Bar
* Hist
* Line
* Subplots

```python
x = np.linspace(0, 10, 100)
x[:10]
```

```python
fig, ax = plt.subplots()
ax.plot(x, x**2);
```

```python
fig, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x));
```

```python
nut_prices = {'almond': 100,
             'peanut': 200,
             'cashew': 300}
fig, ax = plt.subplots()
ax.bar(nut_prices.keys(), nut_prices.values())
ax.set(title='Nut Prices',
      xlabel='Nuts',
      ylabel='Prices')
plt.show()
```

```python
a = np.random.randn(1000)
fig, ax = plt.subplots()
ax.barh(list(nut_prices.keys()), list(nut_prices.values()));
```

```python
fig, ax = plt.subplots()
ax.hist(a);
```
